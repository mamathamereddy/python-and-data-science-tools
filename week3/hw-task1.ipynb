```python
import numpy as np
import pandas as pd
```

```python
#load the data--read in the data and print it
df = pd.read_csv('responses.csv')
df
```

```python
#identify nonsensical values--deleting minSekRaw column
df1=df.drop(columns=['minSekRaw'])
df1
```

```python
#remove/change any nonsensical values
df1['min'] = pd.to_numeric(df1['min'], errors = 'coerce') # change min from string to numeric
df1
```

```python
df2 = df1.dropna() #removing all NaN values
df2
```

```python
df2['time'] = df2['min']*60 + df2.sec # make a time variable
df2
```

```python
mean_time_greek = df2[df2['type'] == 'Greek'].time.mean()
mean_time_greek
```

```python
mean_time_latin = df2[df2['type'] == 'Latin'].time.mean()
mean_time_latin
```

```python
if  mean_time_latin < mean_time_greek:
  print("Latin participants in the sudoku experiment were faster")

else:
  print("Greek participants in the sudoku experiment were faster")
```
